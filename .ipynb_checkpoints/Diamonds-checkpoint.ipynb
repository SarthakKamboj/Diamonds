But before we continue, it is important to define a few key terms so we understand the data we are working with.

Terminology
1. Carat: The weight measurement for diamonds. One carat is equal to 200 milligrams.
2. Cut: This is the style or design to shape a diamond. Its cut has a massive influence in its luminosity.
3. Color: This is based on how colorless and transparent a diamond is. The best color is "D" while the worst is "J".
4. Depth Percentage: This percentage is calculated by dividing the height of the diamond by the widt hof the diamond.
5. Table Percentage: The table is the flat table-like surface on the top of a diamond. The table percentage is the width of the table divided by the width of the entire diamond.
6. Clarity: The quality of the diamond related to the number of inclusion it contains. It is usually represented with labels from FL to I3.
7. Price: The price of the diamond.
8. X: Length of the diamond in millimeters.
9. Y: Width of the diamond in millimeters.
10. Z: Depth of the diamond in millimeters.

<p>
    Lastly, diamond color is on a spectrum from "D" to "J", where "D" is the best and "J" is the worst. Although these letters would not be too diffcult to understand, for simplicity, I will convert them to integers to "D" is represented by 6 and "J" is represented by a 0.
</p>

```python
def organize_table_levels(table):
    if table >= 53 and table <= 58:
        return "Excellent"
    elif table >= 52 and table < 53 or table > 58 and table <= 60:
        return "Very Good"
    elif table >= 51 and table < 52 or table > 60 and table <= 64:
        return "Good"
    elif table >= 50 and table < 51 or table > 64 and table <= 69:
        return "Fair"
    else:
        return "Poor"

def organize_depth_levels(depth):
    if depth >= 59 and depth <= 62.4:
        return "Excellent"
    elif depth >= 58 and depth < 59 or depth > 62.4 and depth <= 63.5:
        return "Very Good"
    elif depth >= 57.5 and depth < 58 or depth > 63.5 and depth <= 64.1:
        return "Good"
    elif depth >= 56.5 and depth < 57.5 or depth > 64.1 and depth <= 65:
        return "Fair"
    else:
        return "Poor"
    
def categorize_colors(clarity):
    if clarity == "IF":
        return "Flawless"
    elif clarity == "VVS1" or clarity == "VVS2":
        return "Very Very Slightly Included"
    elif clarity == "VS1" or clarity == "VS2":
        return "Very Slightly Included"
    elif clarity == "SI1" or clarity == "SI2":
        return "Slightly Included"
    elif clarity == "I1":
        return "Included"
    
def color_rank(color):
    return -(ord(color) - ord("J"))
```

```python
dm_df["table levels"] = dm_df["table"].apply(lambda x: organize_table_levels(x))
dm_df["depth levels"] = dm_df["depth"].apply(lambda x: organize_depth_levels(x))
dm_df["clarity quality"] = dm_df["clarity"].apply(lambda x: categorize_colors(x))
dm_df["color rank"] = dm_df["color"].apply(lambda x: color_rank(x))
```

```python
cat_cols.append("table levels")
cat_cols.append("depth levels")
cat_cols.append("clarity quality")
cat_cols.append("color rank")

cat_cols.remove("clarity")
cat_cols.remove("color")
```

```python
dm_df = pd.concat([dm_df[num_cols],dm_df[cat_cols]],axis=1)
dm_df.head()
```

```python
cat_cols
```

Now, our categorical variables consist of cut, color, clarity, table levels, clarity quality, and depth levels.

```python
fig = plt.figure(1,(20,30))
i = 1
for cat in cat_cols:
    ax = plt.subplot(3,2,i)
    ax.set_xlabel(None)
    ax.set_ylabel(None)
    title_dict = {
        "fontsize":20,
    }
    ax.set_title(f"Distribution of {cat}",fontdict=title_dict)
    sns.countplot(dm_df[cat],order=dm_df[cat].value_counts().index)
    for p in ax.patches:
        height = p.get_height()
        ax.text(p.get_x()+p.get_width()/2,height+3,f"{round(height/len(dm_df),8)}",ha="center",fontsize=15)
    plt.tight_layout()
    i+=1
```

Insights
1. Cut is very imbalanced as about 40% of the data is ideal; however, this makes sense as most likely professionals shaped the diamonds.
2. Table levels and depth levels are insanely distorted, as "Excellent" accounts for approximately 70% of the data in both cases and about 90% of the data is either "Excellent" or "Very Good".
3. Clarity quality are also not balanced as the most dominant classes account for 20%-25% of the total data even though these two categories have the most classes.
4. Color ranks seems to be pretty well distributed since 4 of the categories comprise of about 60% of the data. Overall, although the data does not appear to be normal, it is decently distributed amongst the classes.

```python
combos = itertools.combinations(["cut","color rank","clarity quality"],2)
fig = plt.figure(1,(10,30))
i = 1
for combo in combos:
    ax = plt.subplot(3,1,i)
    hue_big_to_small = dm_df[combo[1]].value_counts().index
    sns.countplot(x=combo[0],hue=combo[1],hue_order=hue_big_to_small,data=dm_df)
    ax.set_title(f"{combo[0]} bifurcated by {combo[1]}")
    ax.set_xlabel(None)
    ax.set_ylabel(None)
    plt.tight_layout()
    i+=1
```

Insights
- Cut bifurcated by color rank
    - A fair cut is pretty equally distributed amongst the different color ranks
    - In all cuts, a color rank of 3 is at least in the top 3 most populated because it was the overall dominant class in this category
    - <strong>Overall the distribution of color rank in each cut seems to be similar with the overall distributions, implying the two categories are likely independent</strong>
